# Diabetes Detection Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Importing the course datasets 
diabetes = pd.read_csv('datasets/diabetes_clean.csv')
```

```python
diabetes.head(2)
```

## Try to predict Blood Glucose depending on the BMI (Body Mass Index)

```python
diabetes[['glucose', 'bmi']].head(2)
```

```python
# The feature will be only the column bmi.
features = diabetes['bmi'].values.reshape(-1, 1)
```

```python
# The target will be the glucose
target = diabetes['glucose'].values
```

### Representing the relation between BMI and Glucose in a Graph

```python
# points diagram
plt.scatter(features, target)

plt.xlabel("Body Mass Index")
plt.ylabel("Blood glucose level (ml/dl)")

plt.show()
```

### Creation of a Linear Regression Model

```python
model = LinearRegression()
```

### Training of the model
As we are modeling the relationship between the feature and the target, rather than predicting target values for new observation, we fit the model to all of our observations

```python
model.fit(features, target)
```

### Get the prediction that our model does 

```python
predictions = model.predict(features)
```

```python

print(*predictions[:4])
print(*target[:4])
print(*features[:4])
```

```python
# Show the real value and the predicted one in consecutive columns
# Calculate the difference between the predicted value and the real one
df = pd.DataFrame(list(zip(target, predictions)), 
                  columns =['real', 'predicted'])
df['error'] = df['real'] - df['predicted']
df.head(5)
```

### Draw the line of the Linear Regression

```python
plt.scatter(features, target)
plt.xlabel("BMI")
plt.ylabel("Glucose")

plt.plot(features, predictions, 'C1') # C1 is the color selection
plt.show()
```
